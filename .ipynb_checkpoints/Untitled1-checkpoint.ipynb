# Covid-19 cases in Nepal
## Mapping cases In Map


importing basic requrements

```python
import requests
import folium
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Using publicAPi and using request method and storing data

```python
TestingSummaryResponse=requests.get('https://nepalcorona.info/api/v1/data/nepal')
CaseByMuciplalityResponse=requests.get('https://data.nepalcorona.info/api/v1/covid')
DistrictsList=requests.get('https://data.nepalcorona.info/api/v1/districts')
TimelineResponse=requests.get('https://data.nepalcorona.info/api/v1/covid/timeline')
munciplatyList=requests.get('https://data.nepalcorona.info/api/v1/municipals')
```

```python
#converting to json files
munciplatyLists= munciplatyList.json()
```

```python
#changing json file to list
munciplaty=[]
for detail in munciplatyLists:
    munciplaty.append({'id':detail['id'],'name':detail['code'],'coordinateX':detail[ 'centroid']['coordinates'][0],'coordinateY':detail[ 'centroid']['coordinates'][1]})
   
    
```

for timeline of cases

```python
timeline=TimelineResponse.json()
```

```python
#converting to list
date=[]
newCases=[]
newRecovery=[]
totalDeath=[]
newDeaths=[]
totalCases=[]

for package in timeline:
    date.append(package['date'])
    newCases.append(package['newCases'])
    newRecovery.append(package['newRecoveries'])
    totalDeath.append(package['totalDeaths'])
    newDeaths.append(package['newDeaths'])
    totalCases.append(package['totalCases'])
```

plotting cases using matplotlib

```python
a=0
plt.figure(figsize=(15,15))
plt.plot(date,newCases,label='newCases')
plt.plot(date,newRecovery,label='newRecovery')
plt.plot(date,newDeaths,label='newDeath')
plt.legend()
plt.xlabel('date')
ax=plt.gca()

for label in ax.get_xaxis().get_ticklabels():
    if (a==4):
        
        a=0
    else:
        
        label.set_visible(False)
        a+=1
plt.xticks(rotation=90)
plt.ylabel('total Cases')
plt.title("Corona cases timeline In context of Nepal")
```

```python
#to plot logarathmic garph, using math.log to change to log
import math
math.log(10)
log=[]
deathlog=[]
for x in newCases:
    if(x==0):
        log.append(0)
    else:
        log.append(math.log(x))

length=len(log)
cu_list=[sum(log[0:x:1]) for x in range (0,length+1)]

for x in newDeaths:
    if(x==0):
        deathlog.append(0)
    else:
        deathlog.append(math.log(x))
length=len(deathlog)
decu_list=[sum(deathlog[0:x:1]) for x in range (0,length+1)]
```

```python

plt.figure(figsize=(20,10))

plt.plot(date,cu_list[1:],label='newCases')
plt.plot(date,decu_list[1:],label='death')
plt.xlabel('date')
ax=plt.gca()
a=0
for label in ax.get_xaxis().get_ticklabels():
    if (a==4):
        
        a=0
    else:
        
        label.set_visible(False)
        a+=1
        
plt.xticks(rotation=90)
plt.yscale('log')
plt.grid(True)
```
